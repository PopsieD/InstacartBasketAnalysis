# 4.

```python
#create loyalty flag

df['max_order']= df.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#view head(100)
pd.options.display.max_rows= None
df.head(100)
```

```python
df.loc[df['max_order']>40, 'loyalty_flag']= 'Loyal customer'
df.loc[(df['max_order']<=40) & (df['max_order']>10), 'loyalty_flag']= 'Regular customer'
df.loc[df['max_order']<=10, 'loyalty_flag']= 'New customer'
```

```python
#view value_counts
df['loyalty_flag'].value_counts(dropna= False)
```

```python
#view relevant columns
df[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

```python
df['spending_category'] = df.groupby(['user_id'])['order_number'].transform('sum')
```

```python
df.loc[df['spending_category']>=10, 'spending_flag']= 'High spender'
df.loc[df['spending_category']<10, 'spending_flag']= 'Low spender'
```

```python
#view value_counts
df['spending_flag'].value_counts(dropna= False)
```

```python
#view relevant columns
df[['user_id', 'spending_flag']].head(60)
```

# 7.

```python
#group and find median
df['behavior_category'] = df.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
df['behavior_category'].head(30)
```

```python
#having issues with the data type not being an integer
df['behavior_category']=df['behavior_category'].astype('int64')
```

```python
#apply values to flag column
df.loc[df['behavior_category']>20, 'behavior_flag']= 'Non-frequent customer'
```

```python
df.loc[(df['behavior_category']<=20) & (df['behavior_category']>10), 'behavior_flag']= 'Regular customer'
```

```python
df.loc[df['behavior_category']<=10, 'behavior_flag']= 'Frequent customer'
```

```python
#view the number of values in each column
df['behavior_flag'].value_counts(dropna= False)
```
